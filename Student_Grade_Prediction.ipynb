```python
df['GradeAvg'] = (df['G1'] + df['G2'] + df['G3']) / 3
```

```python
df.drop(["school","age"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df_dum=df
```

```python
#Converting to categorical value
categorical_d = {'yes': 1, 'no': 0}
df_dum['schoolsup'] = df_dum['schoolsup'].map(categorical_d)
df_dum['famsup'] = df_dum['famsup'].map(categorical_d)
df_dum['paid'] = df_dum['paid'].map(categorical_d)
df_dum['activities'] = df_dum['activities'].map(categorical_d)
df_dum['nursery'] = df_dum['nursery'].map(categorical_d)
df_dum['higher'] = df_dum['higher'].map(categorical_d)
df_dum['internet'] = df_dum['internet'].map(categorical_d)
df_dum['romantic'] = df_dum['romantic'].map(categorical_d)
```

```python
categorical_d = {'F': 1, 'M': 0}
df_dum['sex'] = df_dum['sex'].map(categorical_d)

# map the address data
categorical_d = {'U': 1, 'R': 0}
df_dum['address'] = df_dum['address'].map(categorical_d)

# map the famili size data
categorical_d = {'LE3': 1, 'GT3': 0}
df_dum['famsize'] = df_dum['famsize'].map(categorical_d)

# map the parent's status
categorical_d= {'T': 1, 'A': 0}
df_dum['Pstatus'] = df_dum['Pstatus'].map(categorical_d)

# map the parent's job
categorical_d = {'teacher': 0, 'health': 1, 'services': 2,'at_home': 3,'other': 4}
df_dum['Mjob'] = df_dum['Mjob'].map(categorical_d)
df_dum['Fjob'] = df_dum['Fjob'].map(categorical_d)

# map the reason data
categorical_d= {'home': 0, 'reputation': 1, 'course': 2,'other': 3}
df_dum['reason'] = df_dum['reason'].map(categorical_d)

# map the guardian data
categorical_d = {'mother': 0, 'father': 1, 'other': 2}
df_dum['guardian'] = df_dum['guardian'].map(categorical_d)
```

```python
df_dum.columns
```

```python
from sklearn.model_selection import train_test_split
x=df_dum.drop("G3",axis=1)
y=df_dum['G3']
```

```python
df_dum['G3']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.20, random_state=44)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
L=LinearRegression()
```

```python
L.fit(X_train, y_train)
```

```python
y_pred=L.predict(X_test)
```

```python
print(L.score(X_test, y_test)) 
```

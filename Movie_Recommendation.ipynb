```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df= df.dropna(subset=['Genre', 'Director', 'Series_Title'])
```

```python
df['Features'] = df['Genre'] + " " + df['Director']
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Features'])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend_movies(title, cosine_sim=cosine_sim, data=df):
    # Find the index of the given movie title
    idx = df[df['Series_Title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_indices = [i[0] for i in sim_scores[1:11]]
    return data['Series_Title'].iloc[sim_indices]
```

```python
movie_name = "The Godfather"  # Replace with a movie name in the dataset
recommended_movies = recommend_movies(movie_name)

print(f"Movies similar to '{movie_name}':")
print(recommended_movies)
```
